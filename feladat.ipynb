```python
# √Årak helyes t√≠pusba konvert√°l√°sa
df['price'] = df['price'].str.replace('[\$,]', '', regex=True).astype(float)
```

```python
df.dtypes
```

**Mik az √°rak sz√©ls≈ë√©rt√©kei, √°tlaga, medi√°nja, sz√≥r√°sa?**

```python
# √Årak sz√©ls≈ë√©rt√©keinek meghat√°roz√°sa
min_price = df['price'].min()
max_price = df['price'].max()

print(f"A legalacsonyabb √°r: {min_price} eur√≥")
print(f"A legmagasabb √°r: {max_price} eur√≥")
```

```python
# √Årak √°tlag√°nak meghat√°roz√°sa
mean_price = df['price'].mean()

print(f"Az √°rak √°tlaga: {mean_price:.2f} eur√≥")
```

```python
# √Årak medi√°nj√°nak meghat√°roz√°sa
median_price = df['price'].median()

print(f"Az √°rak medi√°nja: {median_price:.2f} eur√≥")
```

```python
# √Årak sz√≥r√°s√°nak meghat√°roz√°sa
std_price = df['price'].std()

print(f"Az √°rak sz√≥r√°sa: {std_price:.2f} eur√≥")
```

**H√°nyf√©le CPU architekt√∫ra tal√°lhat√≥ meg az adathalmazban?**

```python
# H√°nyf√©le CPU architekt√∫ra tal√°lhat√≥ meg az adathalmazban?
cpu_architectures = df['cpu'].unique()
num_architectures = len(cpu_architectures)

print(f"Az adathalmazban {num_architectures} k√ºl√∂nb√∂z≈ë CPU architekt√∫ra tal√°lhat√≥.")
```

```python
# Csoportos√≠t√°s CPU architekt√∫ra alapj√°n √©s statisztik√°k sz√°m√≠t√°sa
cpu_price_stats = df.groupby('cpu')['price'].agg(['min', 'max', 'mean', 'median', 'std'])

cpu_price_stats
```
