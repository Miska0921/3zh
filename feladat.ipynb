Line√°ris modell seg√≠ts√©g√©vel megj√≥solhat√≥-e egyed√ºl a harddisk m√©ret √©s a ram m√©rete alapj√°n a k√©sz√ºl√©k √°ra? Jav√≠that√≥-e a becsl√©s ha figyelembe vesz√ºnk m√°s

```python
# Sz√ºks√©ges k√∂nyvt√°rak import√°l√°sa a line√°ris feladathoz
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
# A harddisk oszlop konvert√°l√°sa GB-ra m√°s √©rt√©kekr≈ël
df['harddisk_numeric'] = df['harddisk'].str.upper()
df['harddisk_numeric'] = df['harddisk_numeric'].str.replace('TB', '000GB')
df['harddisk_numeric'] = df['harddisk_numeric'].str.replace('GB', '')
df['harddisk_numeric'] = pd.to_numeric(df['harddisk_numeric'], errors='coerce')
```

```python
# Hi√°nyz√≥ harddisk √©rt√©kek kezel√©se medi√°nnal
harddisk_median = df['harddisk_numeric'].median()
df['harddisk_numeric'] = df['harddisk_numeric'].fillna(harddisk_median)
```

```python
# A ram oszlop konvert√°l√°sa numerikusra
df['ram_numeric'] = df['ram'].str.replace('GB', '')
df['ram_numeric'] = pd.to_numeric(df['ram_numeric'], errors='coerce')
```

```python
# Hi√°nyz√≥ ram √©rt√©kek kezel√©se medi√°nnal
ram_median = df['ram_numeric'].median()
df['ram_numeric'] = df['ram_numeric'].fillna(ram_median)
```

```python
# Hi√°nyz√≥ √°rak kezel√©se medi√°nnal
price_median = df['price'].median()
df['price'] = df['price'].fillna(price_median)
```

```python
df.head(10)
```

```python
# 1. Egyszer≈± lin√°ris modell: csak harddisk √©s ram felhaszn√°l√°s√°val
X_simple = df[['harddisk_numeric', 'ram_numeric']].values
y = df['price'].values

model_simple = LinearRegression()
model_simple.fit(X_simple, y)

y_pred_simple = model_simple.predict(X_simple)
r2_simple = r2_score(y, y_pred_simple)
mse_simple = mean_squared_error(y, y_pred_simple)

print("Egyszer≈± modell (csak harddisk √©s ram):")
print(f"R¬≤ √©rt√©k: {r2_simple:.4f}")
print(f"MSE: {mse_simple:.2f}")
print(f"Koefficiensek: harddisk = {model_simple.coef_[0]:.2f}, ram = {model_simple.coef_[1]:.2f}")
print(f"Intercept: {model_simple.intercept_:.2f}\n")
```

```python
# 2. B≈ëv√≠tett modell: harddisk, ram √©s cpu
# CPU k√≥dol√°sa
df['cpu'] = df['cpu'].fillna('Unknown')
label_encoder = LabelEncoder()
df['cpu_encoded'] = label_encoder.fit_transform(df['cpu'])
```

```python
X_extended = df[['harddisk_numeric', 'ram_numeric', 'cpu_encoded']].values

model_extended = LinearRegression()
model_extended.fit(X_extended, y)
```

```python
y_pred_extended = model_extended.predict(X_extended)
r2_extended = r2_score(y, y_pred_extended)
mse_extended = mean_squared_error(y, y_pred_extended)
```

```python
print("B≈ëv√≠tett modell (harddisk, ram, cpu):")
print(f"R¬≤ √©rt√©k: {r2_extended:.4f}")
print(f"MSE: {mse_extended:.2f}")
print(f"Koefficiensek: harddisk = {model_extended.coef_[0]:.2f}, ram = {model_extended.coef_[1]:.2f}, cpu = {model_extended.coef_[2]:.2f}")
print(f"Intercept: {model_extended.intercept_:.2f}\n")
```

```python
# Eredm√©nyek √∂sszehasonl√≠t√°sa
print("√ñsszehasonl√≠t√°s:")
print(f"R¬≤ javul√°s: {(r2_extended - r2_simple):.4f}")
print(f"MSE cs√∂kken√©s: {(mse_simple - mse_extended):.2f}")
if r2_extended > r2_simple:
    print("A b≈ëv√≠tett modell jobb teljes√≠tm√©nyt ny√∫jt, mivel az R¬≤ √©rt√©k n≈ëtt √©s az MSE cs√∂kkent.")
else:
    print("A b≈ëv√≠tett modell nem jav√≠tott jelent≈ësen a teljes√≠tm√©nyen.")
```
