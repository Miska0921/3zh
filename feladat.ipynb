```python
# Adatok beolvasása
import pandas as pd
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/laptop_prices.csv')
#df = pd.read_csv('laptop_prices.csv')
```

```python
# Árak helyes típusba konvertálása
df['price'] = df['price'].str.replace('[\$,]', '', regex=True).astype(float)
```

```python
df.dtypes
```

**Mik az árak szélsőértékei, átlaga, mediánja, szórása?**

```python
# Árak szélsőértékeinek meghatározása
min_price = df['price'].min()
max_price = df['price'].max()

print(f"A legalacsonyabb ár: {min_price} euró")
print(f"A legmagasabb ár: {max_price} euró")
```

```python
# Árak átlagának meghatározása
mean_price = df['price'].mean()

print(f"Az árak átlaga: {mean_price:.2f} euró")
```

```python
# Árak mediánjának meghatározása
median_price = df['price'].median()

print(f"Az árak mediánja: {median_price:.2f} euró")
```

```python
# Árak szórásának meghatározása
std_price = df['price'].std()

print(f"Az árak szórása: {std_price:.2f} euró")
```

**Hányféle CPU architektúra található meg az adathalmazban?**

```python
# Hányféle CPU architektúra található meg az adathalmazban?
cpu_architectures = df['cpu'].unique()
num_architectures = len(cpu_architectures)

print(f"Az adathalmazban {num_architectures} különböző CPU architektúra található.")
```

**Mik az árak szélsőértékei, átlaga, mediánja, szórása a különböző cpu architektúrákra nézve?**

```python
# Csoportosítás CPU architektúra alapján és statisztikák számítása
cpu_price_stats = df.groupby('cpu')['price'].agg(['min', 'max', 'mean', 'median', 'std'])

cpu_price_stats
```
